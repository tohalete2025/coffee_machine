```python
# Exploratory Data Analysis
```

```python
df.pivot_table(values='Sales', index='City', columns='Date', aggfunc='sum')
```

```python
import seaborn as sns
```

```python
sns.histplot(df['Sales'])
```

```python
sns.boxplot(x='City', y='Sales', data=df)
```

```python
# Total sales per Branch
branch_sales = df.groupby('Branch')['Sales'].sum()
print(branch_sales)
```

```python
# Total sales per Product Line
product_sales = df.groupby('Product line')['Sales'].sum()
print(product_sales)
```

```python
# Total sales per Payment Method
payment_method = df.groupby('Payment')['Sales'].sum()
print(payment_method)
```

```python
# Total sales per Customer type
customer_patronage = df.groupby('Customer type')['Sales'].sum()
print(customer_patronage)
```

```python
# Visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Barplot of total sales by region
sns.barplot(x=branch_sales.index, y=branch_sales.values)
plt.title('Sales by Branch')
plt.ylabel('Sales')
plt.show()
```

```python
sns.barplot(x=product_sales.index, y=product_sales.values, color='red')
plt.title('Sales by Product Line')
plt.ylabel('Sales')
plt.show()
```

```python
sns.barplot(x=payment_method.index, y=payment_method.values, color='red')
plt.title('Sales by Payment Method')
plt.ylabel('Sales')
plt.show()
```

```python
sns.barplot(x=customer_patronage.index, y=customer_patronage.values, color='green')
plt.title('Sales by membership')
plt.ylabel('Sales')
plt.show()
```

```python
# Boxplot of Sales
sns.boxplot(x='Customer type', y='Sales', data=df)
plt.show()
```

```python
# Correlation heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

# Summary of Insights from Supermarket Analysis:

# 1. Sales by Branch, Product Line and Payment method are almost all identical
# as there appears to be no much difference

# 2. Sales recorded for membership is significantly more than that recorded for
# non members or normal customers

# 3. Surpisingly average Rating by non-members or normal customers is slightlty higher
# than Rating by members. This may imply that with the right promotional activities in
# place, there is a high likelihood of converting non-members to members and getting
